```python
import pandas as pd

import sklearn.cluster as cls
from sklearn.cluster import KMeans

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('CityGroup.csv')
```

```python
df
```

```python
x = df['Elapsed_Time']
y = df['Age']

df2 = df[['Elapsed_Time', 'Age']]
kMeans = cls.KMeans(n_clusters=2)
kMeans.fit(df2)
```

```python
centroids = kMeans.cluster_centers_
print(centroids)
```

```python
centroid1 = centroids[:, 0]
centroid2 = centroids[:, 1]
```

```python
label=kMeans.labels_
print(label)
```

```python
x = df['Elapsed_Time']
y = df['Age']
plot1= x
plot2= y
plt.figure(figsize=(10, 5))

plt.scatter(x, y, c=label, cmap='rainbow')
plt.scatter(centroid1, centroid2, marker='X', color='black', s=300)

plt.title('City Group', size=15)
plt.xlabel('Elapsed Time', size=12)
plt.ylabel('Age', size=12)

plt.savefig('City Group.png')

plt.show()
```
